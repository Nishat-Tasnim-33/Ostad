```python
"""1. Read a CSV file 
Here the CSV contains tain dataset of WIELDCATS 
"""
import pandas as pd
```

```python
read = pd.read_csv('(Titanic)train.csv')
print(read.head())
```

```python
"""Checking shape of the dataset"""
read2 = read.shape
print("The shape of the dataset is",read2)
read3 = read.isnull().sum().sum()
print("The number of null values in the column is ", read3)
```

```python
"""2. Find the duplicates & replace them"""
read5 = read.duplicated()
print(read5.to_string())
```

```python
read5.drop_duplicates()
###As in the data set there is no duplicate so nothing is returned.
```

```python
"""3. Check missing values"""
print(read.isnull().to_string())
```

```python
"""4.Filling the missing value with any constrant value"""

read = read.fillna(value=-1)
print(read)
```

```python
"""5. Group by mean, median, sum"""
read.groupby('Survived').sum()
```

```python
read.groupby('Survived').mean()
```

```python
read.groupby('Survived').median()
```
